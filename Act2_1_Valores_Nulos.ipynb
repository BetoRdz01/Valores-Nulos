```python
import pandas as pd
```

```python
df = pd.read_csv('Ventas_totales.csv')
pd.options.display.float_format = '{:.2f}'.format
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df2=df.copy()
```

```python
#salon_ventas, reemplazamos con la media dado a que son valores numéricos y considero que la media representa bien las ventas en el salón
df2['salon_ventas']=df2['salon_ventas'].fillna(round(df['salon_ventas'].mean(),1)) 

#Reemplazamos valores en tarjetas de débito y crédito con la mediana, siendo cantidades que pueden expresarse con esta métrica, sin tener influencia de valores atípicos.  
df2['tarjetas_debito']=df2['tarjetas_debito'].fillna(round(df['tarjetas_debito'].median(),1))
df2['tarjetas_credito']=df2['tarjetas_credito'].fillna(round(df['tarjetas_credito'].median(),1))

#Reemplazamos con la media para los otros medios y el subtotal de ventas en alimentos y bebidas, pues se considera que la media puede representar correctamente estos pocos
#valores faltantes siendo una medida de tendencia central
df2['otros_medios']=df2['otros_medios'].fillna(round(df['otros_medios'].mean(),1))
df2['subtotal_ventas_alimentos_bebidas']=df2['subtotal_ventas_alimentos_bebidas'].fillna(round(df['subtotal_ventas_alimentos_bebidas'].mean(),1))
```

```python
df2.isnull().sum()
```

```python
#Llenamos nulos con datos del día anterior (forward fill), siendo un día que probablemente tiene un comportamiento similar.
df2['bebidas']=df2['bebidas'].fillna(method='ffill')
df2['almacen']=df2['almacen'].fillna(method='ffill')
df2['panaderia']=df2['panaderia'].fillna(method='ffill')
```

```python
df2.isnull().sum()
```

```python
#Similar a lo mostrado anteriormente, utilizamos bfill para llenar los datos nulos con los registros del siguiente día. Esperando que tenga un comportamiento similar. 
df2['lacteos']=df2['lacteos'].fillna(method='bfill')
df2['carnes']=df2['carnes'].fillna(method='bfill')
df2['verduleria_fruteria']=df2['verduleria_fruteria'].fillna(method='bfill')
df2['alimentos_preparados_rotiseria']=df2['alimentos_preparados_rotiseria'].fillna(method='bfill')
```

```python
df2.isnull().sum()
```

```python
#Obtenemos que percentil 50 de la columna
df2[['indumentaria_calzado_textiles_hogar', 'electronicos_articulos_hogar', 'otros']].describe()
```

```python
#reemplazamos los nulos de las columnasrestantes con el percentil 50, una métrica de posición que identifica adecuadamente lo que sería la posición del 50% de los datos.  
df2['indumentaria_calzado_textiles_hogar']=df2['indumentaria_calzado_textiles_hogar'].fillna(1426312.60)
df2['electronicos_articulos_hogar']=df2['electronicos_articulos_hogar'].fillna(3215054.67)
df2['otros']=df2['otros'].fillna(4634783.45)
```

```python
df2.isnull().sum()
```
